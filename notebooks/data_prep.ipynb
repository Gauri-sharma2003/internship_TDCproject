```python
import matplotlib.pyplot as plt
def remove_outliers_and_plot(df, columns):
    print("_"*30)
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))
    print("Shape Before Removing Outlier : ",df.shape)
    # Draw box plot before removing outliers
    df.boxplot(column=columns, ax=axes[0])
    axes[0].set_title('Before Removing Outliers')
    for column in columns:
        
        
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

    # Draw box plot after removing outliers
    df.boxplot(column=columns, ax=axes[1])
    axes[1].set_title('After Removing Outliers')

    plt.show()
    print("Shape After Removing Outlier : ",df.shape)
    print("_"*30)
    return df
```

```python
data_ro = pd.DataFrame(columns=data_rn.columns)
for application in data_rn['Application'].unique():
    print("Application : ",application)
    df_ro_temp = remove_outliers_and_plot(data_rn[data_rn['Application']==application],Outlier_Columns)
    data_ro =pd.concat([data_ro, df_ro_temp], ignore_index=True)
```

```python
import seaborn as sns
def draw_normal_distribution(df, columns, hue_column):
    for column in columns:
        fig, ax = plt.subplots(figsize=(8, 6))
        sns.histplot(data=df, x=column, hue =df[hue_column],kde=True, ax=ax, stat="density", common_norm=False)
        ax.set_title(f'Normal Distribution for {column}')
        ax.set_xlabel(column)
        ax.set_ylabel('Density')
        ax.legend(df[hue_column].unique())
        plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
def normalize_data(df):
    # Separate numerical and categorical columns
    numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
    categorical_columns = df.select_dtypes(include=['object']).columns

    # Initialize MinMaxScaler
    scaler = MinMaxScaler()
    
    # Fit the scaler to the numerical data and transform it
    df_normalized = df.copy()
    df_normalized[numerical_columns] = scaler.fit_transform(df[numerical_columns])
    
    return df_normalized
```

```python
def clean_data(data, label_encode_cols=[], onehot_encode_cols=[]):
    # Handle missing values
    data.fillna(0, inplace=True)  # Example: fill missing values with 0
    
    # Encode categorical variables using LabelEncoder
    for col in label_encode_cols:
        data[col] = LabelEncoder().fit_transform(data[col])
    
    # Encode categorical variables using OneHotEncoder
    for col in onehot_encode_cols:
        encoded_cols = pd.get_dummies(data[col], prefix=col,drop_first=True)
        encoded_cols = encoded_cols.astype(int)
        data = pd.concat([data, encoded_cols], axis=1)
        data.drop(col, axis=1, inplace=True)
    
    
    
    
    # Initialize MinMaxScaler
    scaler = MinMaxScaler()
    
    numerical_cols = data.select_dtypes(include=['int', 'float']).columns
    
    # Fit the scaler to the numerical data and transform it
    df_normalized = data.copy()
    df_normalized[numerical_cols] = scaler.fit_transform(data[numerical_cols])
    
    return df_normalized
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
```

```python
def train_and_evaluate_models(X_train, X_test, y_train, y_test):
    models = {
        'Logistic Regression': LogisticRegression(),
        'Decision Tree': DecisionTreeClassifier(),
        'Random Forest': RandomForestClassifier(),
        'SVM': SVC(),
        'K-Nearest Neighbors': KNeighborsClassifier(),
        'Naive Bayes': GaussianNB(),
        'Neural Network': MLPClassifier()
    }
    
    results = {}
    for name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        
#         accuracy = accuracy_score(y_test, y_pred)
#         precision = precision_score(y_test, y_pred)
#         recall = recall_score(y_test, y_pred)
#         f1 = f1_score(y_test, y_pred)
        
        cr = classification_report(y_test, y_pred)
        
        print("_"*30)
        print("Model Name: ",name)
        print("CLassification Report",cr)
        
        results[name]=cr
    
#     results_df = pd.DataFrame(results)
#     best_model = results_df.loc[results_df['F1 Score'].idxmax()]
    
    return results
```

```python
results = train_and_evaluate_models(X_train, X_test, y_train, y_test)
```

```python
def classification_report_to_dataframe(classification_report_str):
    report_dict = classification_report_str.split('\n')
    report_data = [row.split() for row in report_dict[2:-5]]

    df = pd.DataFrame(report_data, columns=['class', 'precision', 'recall', 'f1-score', 'support'])
    df.set_index('class', inplace=True)
    return df
```
